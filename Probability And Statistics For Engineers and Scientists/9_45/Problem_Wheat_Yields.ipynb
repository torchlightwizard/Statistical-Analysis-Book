**Problem Statement:** The federal government awarded grants to the
agricultural departments of 9 universities to test the
yield capabilities of two new varieties of wheat. Each
variety was planted on a plot of equal area at each
university, and the yields, in kilograms per plot.

**Objective:** Finding the mean difference between the yields of variety 1 and variety 2 with a 95% confidence interval.

```python
import pandas as pd

data = pd.read_csv("Data_9_45.csv", header=None)
data
```

```python
data = data.T
variety1 = data[0]
variety2 = data[1]
```

**Method 1:** If population variances are not known and are assumed to be different.

```python
import numpy as np

x1 = np.mean(variety1)
x2 = np.mean(variety2)
s1 = np.std(variety1, ddof=1)
s2 = np.std(variety2, ddof=1)
n1 = variety1.shape[0]
n2 = variety2.shape[0]
print(x1, x2, s1, s2, n1, n2)
```

```python
import math

v = (s1**2/n1 + s2**2/n2)**2 / ((s1**2/n1)**2/(n1-1) + (s2**2/n2)**2/(n2-1)) # degree of freedom
v = math.floor(v)
t_val1 = 2.131
x_d1 = abs(x1 - x2)
e = (t_val1)*(math.sqrt((s1**2/n1) + (s2**2/n2))) # error
ci = (x_d1 - e, x_d1 + e) # confidence interval
print(e)
print(ci)
```

**Method 2:** If we pair the two populations by grouping them by university.

```python
varietyd = variety1 - variety2
xd = abs(np.mean(varietyd))
sd = np.std(varietyd, ddof=1)
nd = varietyd.shape[0]
print(xd, sd, nd)
```

```python
t_val2 = 2.306
e2 = (t_val2)*(sd)*math.sqrt(1/nd)
ci2 = (xd - e2, xd + e2)
print(e2)
print(ci2)
```

**Conclusion:** It is apparent, by using the pair method, our degree of freedom decreased, however our variance and error also decreased. We have a better confidence interval by pairing the data as compared to keeping the populations seperate.[link text](https://)
